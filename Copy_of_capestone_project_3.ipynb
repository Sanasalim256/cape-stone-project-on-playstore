#### Chart - 4

````python
# Chart - 4 visualization code
import numpy as np
import pandas as pd
import seaborn as sns

# ... your existing code ...

# Convert 'Installs' and 'Reviews' to numeric, handling non-numeric values
Installs = pd.to_numeric(df['Installs'].str.replace(',', '').str.replace('+', ''), errors='coerce')
Reviews = pd.to_numeric(df['Reviews'], errors='coerce')

# Apply log transformation after converting to numeric
p = sns.pairplot(pd.DataFrame(list(zip('Rating', 'Size', np.log(Installs), np.log10(Reviews), 'Price', 'Type')),
                        columns=['Rating','Size', 'Installs', 'Reviews', 'Price','Type']), hue='Type')
p.fig.suptitle("Pairwise Plot - Rating, Size, Installs, Reviews, Price",x=0.5, y=1.0, fontsize=16)
````

##### 1. Why did you pick the specific chart?

Answer Here.

The goal is to visualize relationships between multiple variables: Rating, Size, Installs, Reviews, Price, and Type. A pairplot in seaborn is particularly well-suited for this scenario.
Dataset consists of numerical variables (Rating, Size, Installs, Reviews, Price) and a categorical variable (Type). The objective is to explore the relationships between these variables, making pairplot a suitable choice.

##### 2. What is/are the insight(s) found from the chart?

Answer Here
General Insights and Potential Patterns:

1.Rating and Reviews: You might observe a positive correlation between Rating and Reviews. Apps with higher ratings tend to have more reviews, indicating user satisfaction and popularity.

2.Rating and Installs: There could be a positive relationship between Rating and Installs as well. Higher-rated apps are often more likely to be downloaded, although other factors like app visibility and marketing also play a role.

3.Size and Installs: You might find a slight negative correlation between Size and Installs. Larger app sizes can deter users due to storage constraints or download times. However, this relationship can vary depending on the app category and functionality.

4.Price and Installs: Paid apps generally have fewer installs compared to free apps. This is expected as users are more hesitant to pay for apps unless they offer significant value or unique features.
